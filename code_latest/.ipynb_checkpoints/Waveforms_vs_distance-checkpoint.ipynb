```python
## plot distance vs event time
## draw waveform coloured 
## calc S/N
def get_max_ampl(st,ev_time,t_SKS,t_SKKS):
    st_tmp = st
    twin = 120

    timevec=np.linspace(float(st_tmp.stats.starttime),float(st_tmp.stats.starttime)+3600,st_tmp.stats.npts)
    print(UTCDateTime(timevec[0]))
    print(st_tmp.stats.starttime)
    
    timelag = timevec[0]-float(st_tmp.stats.starttime)
    timevec = timevec-timelag
    
    idx_max_ampl = np.where((timevec>float(ev_time)+t_SKS-twin) & (timevec<float(ev_time)+t_SKS+twin))
    max_ampl = np.max(abs(st_tmp.data[idx_max_ampl]))

    idx_ampl = abs(st_tmp.data[idx_max_ampl]).argmax()
    sign = np.sign(st_tmp[idx_max_ampl][idx_ampl])    

    nsec = 7 
    noise_window=45
    NOISE = np.mean(abs(st_tmp.data[idx_max_ampl][0:noise_window]))
    SNR = max_ampl/NOISE
    print(float(timevec[idx_max_ampl][idx_ampl])-float(ev_time))
    print(t_SKS)

    plt.figure(figsize=(16,8))
    plt.plot(timevec,st_tmp.data,'k')    
    plt.plot(timevec[idx_max_ampl][idx_ampl-10*nsec:idx_ampl+10*nsec],st_tmp.data[idx_max_ampl][idx_ampl-10*nsec:idx_ampl+10*nsec],'g',label='S-Phase')
    plt.vlines(float(ev_time)+t_SKS,-max_ampl,max_ampl,color='red',label='SKS-onset')
    plt.vlines(float(ev_time)+t_SKKS,-max_ampl,max_ampl,color='blue',label='SKKS-onset')    
    plt.fill_between([timevec[idx_max_ampl][0],timevec[idx_max_ampl][noise_window*10]],[-NOISE,-NOISE],[max_ampl,max_ampl],alpha=0.5,color='lightgray',label='Noise Window')
    plt.xlim(float(ev_time)+t_SKS-twin,float(ev_time)+t_SKS+twin)
    plt.legend()
    plt.grid()
    plt.title('{0}, {1}, SNR={2}, twin={3} s'.format(st_tmp.stats.station,ev_time.strftime('%Y-%m-%d, %H:%M:%S'),round(SNR,0),round(twin,2)))
    plt.savefig('/media/hein/home2/SplitWave_Results/Splitws/{0}_{1}.png'.format(st_tmp.stats.station,ev_time.strftime('%Y-%m-%d, %H:%M:%S')))
    plt.show()
    return timevec[idx_max_ampl][idx_ampl],max_ampl

#plt.plot(st_tmp.data)
nstat = 9
arr_t,max_ampl = get_max_ampl(st_Z[nstat],ev_time,t_SKS_arr[nstat],t_SKKS_arr[nstat])

```
